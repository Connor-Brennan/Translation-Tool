```python
def corpus_translation():

    #Change workbook as needed for program
    #Change all addresses as needed for program
    cl = load_workbook(r"C:\Users\cmbro\Documents\Modi Speeches NM\NM Corpus4.xlsx")
    ws = cl.active
    i=0
    tr = 0

    #Input the two languages you are translating between here
    languages = ['en', 'hi']

    #Alter min_row, max_row, min_col, max_col for the spreadsheet corpus of documents to be translated
    #The idea is to have a column including every filename and let the program go down the line grabbing them
    for row in ws.iter_rows(min_row=2, min_col=3, max_col=3, max_row=1490):
      for cell in row:
        filename = cell.value + '.txt'
        dets =[]
        print(i)
        if ws.cell(row = 2+i, column = 4).value==1 and ws.cell(row = 2+i, column = 7).value==0 and ws.cell(row = 2+i, column = 10).value == 0:
            print("Examining file.")
            try:
                with open(r"C:\Users\cmbro\Documents\Modi Speeches NM\NM\Modi_speeches_updated\speeches_NM\Buffer Test\Need Translation\{}".format(filename), "r", buffering=1, encoding='utf8') as o:
                    with open(r"C:\Users\cmbro\Documents\Modi Speeches NM\NM\Modi_speeches_updated\speeches_NM\Buffer Test\Holding\R_{}".format(filename), "w+", encoding='utf8') as d:
                        data = o.readlines()
                        #change dest to whichegver language you want to end up with
                        result = translator.translate(data, dest='en')
                        d.write("\n".join(map(lambda x: x.text, result)))
                        print(filename)

                #detect language(s) of each line of file after translation
                with open(r"C:\Users\cmbro\Documents\Modi Speeches NM\NM\Modi_speeches_updated\speeches_NM\Buffer Test\Holding\R_{}".format(filename), "r", encoding='utf8') as d:
                    contents = d.readlines()
                    for line in contents:
                        try:
                            det = detect(line)
                            dets.append(det)
                        except:
                            pass

                print(dets)
                res = list(filter(lambda x: x in languages, dets))
                print(res)

                #If all lines of the file are written in the target language, we file it to a done folder
                #If any line of the file is still written in the old langauge, we file the version with whatever translated
                ##to another folder to run again
                #Alter language of if statement to be langauge translating from
                if 'hi' in res: #It didn't work!
                    print("This one needs help!")
                    os.remove(r"C:\Users\cmbro\Documents\Modi Speeches NM\NM\Modi_speeches_updated\speeches_NM\Buffer Test\Need Translation\{}".format(filename))
                    with open(r"C:\Users\cmbro\Documents\Modi Speeches NM\NM\Modi_speeches_updated\speeches_NM\Buffer Test\Need Translation\{}".format(filename), "w+", encoding='utf8') as d:
                        for line in contents:
                            d.write(line)
                        d.close() 
                    print("Go Again.")
                else: #It worked!
                    os.remove(r"C:\Users\cmbro\Documents\Modi Speeches NM\NM\Modi_speeches_updated\speeches_NM\Buffer Test\Need Translation\{}".format(filename))
                    with open(r"C:\Users\cmbro\Documents\Modi Speeches NM\NM\Modi_speeches_updated\speeches_NM\Buffer Test\Worked\R_{}".format(filename), "w+", encoding='utf8') as d:
                        for line in contents:
                            d.write(line)
                        d.close()
                    print("Translation passed.")

            except FileNotFoundError:
                pass

        i += 1
        
    print("Run again for more translations!")
        
```
